```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv("car data.csv")
```

```python
print('\n Seller Types  - ',data['Seller_Type'].unique())
print('\n Transmission types - ',data['Transmission'].unique())
print('\n Fule types - ',data['Fuel_Type'].unique())
```

```python
Seller_Type=pd.get_dummies(data["Seller_Type"],drop_first=True)
```

```python
Transmission=pd.get_dummies(data["Transmission"],drop_first=True)
```

```python
Fuel_Type=pd.get_dummies(data["Fuel_Type"],drop_first=True)
```

```python
data=data.drop(["Fuel_Type","Transmission","Seller_Type","Car_Name"],axis=1)
```

```python
data.head()
```

```python
data=pd.concat([data,Fuel_Type,Transmission,Seller_Type],axis=1)
```

```python
data
```

```python
y=data["Selling_Price"]
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(data)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lo=LinearRegression()
lo.fit(X_train,y_train)
```

```python
lo.score(X_test,y_test)
```
